```python
#  ใส่ N แทน nan ในตัวแปร PEDROWNOTGRNT ,INATTENTIONIND และ SPEEDING
df1['PEDROWNOTGRNT1']=df1['PEDROWNOTGRNT'].fillna('N')
df1['SPEEDING1']=df1['SPEEDING'].fillna('N')
df1['INATTENTIONIND1']=df1['INATTENTIONIND'].fillna('N')
```

```python
import datetime
def date_sep(date_time):
    p=date_time.split()
    if len(p)==3:
        p1=p[0].split('/')
        q=p[1].split(':')
#         print(q[0],q[1],q[2])
        da_time=datetime.datetime(int(p1[2]),int(p1[0]),int(p1[1]),int(q[0]),int(q[1]),int(q[2]))
#         print(da_time)
        hr_num=int(q[0])
        day_num=int(p1[1])
        year_num=int(p1[2])
        day=da_time.strftime("%A")
        month=da_time.strftime("%B")
    else :
        p1=p[0].split('/')
        q='nan'
        da_time=datetime.datetime(int(p1[2]),int(p1[0]),int(p1[1]))
#         print(da_time)
        hr_num='nan'
        day_num=int(p1[1])
        year_num=int(p1[2])
        day=da_time.strftime("%A")
        month=da_time.strftime("%B")
    return da_time, hr_num,day_num,day,month, year_num
```

```python
#แสดงรายละเอียดวัน,เวลาและปีที่เกิดเหตุ
round1 = lambda x: date_sep(x)
pr= pd.DataFrame(list(df1['INCDTTM'].apply(round1)),columns=['datime','hr_num','day_num','day','month',' year_num'])
pr.head()
```

```python
# รวมข้อมูลระหว่าง df1 และ pr
df=pd.concat([df1,pr],axis=1)
df.head()
# df.to_csv('Clean_Data-Collisions.csv')
```

```python
#  แสดงตัวแปรในตาราง df
df.columns
```

```python
def param_graph(param):
    
    df[param].unique()
    dfg=pd.DataFrame(list(df[param].unique()),columns=[param])
    dfg=dfg.dropna()
    
#     จำนวนข้อมูลของพารามิเตอร์แยกโดยความรุนแรงของการชนกัน
    acc2 = df.groupby([param,'SEVERITYCODE'])['SEVERITYCODE'].count()
    
#     จำนวนข้อมูลของพารามิเตอร์แยกโดยความรุนแรงของการชนกันระดับ 1
    ser_1=acc2[:,1].reset_index()
    pg1=pd.merge(dfg,  
                     ser_1,  
                     on =param,  
                     how ='left') 
    pg1['SEVERITYCODE_1']=pg1['SEVERITYCODE']
    pg1.drop(['SEVERITYCODE'], axis=1, inplace=True)
    
#     จำนวนข้อมูลของพารามิเตอร์แยกโดยความรุนแรงของการชนกันระดับ 2
    ser_2=acc2[:,2].reset_index()
    ser_2['SEVERITYCODE_2']=ser_2['SEVERITYCODE']
    ser_2.drop(['SEVERITYCODE'], axis=1, inplace=True)
    pg2=pd.merge(pg1,  
                     ser_2,  
                     on =param,  
                     how ='left') 
    pg2.dropna(subset=['SEVERITYCODE_1', 'SEVERITYCODE_2'])
    pg2=pg2.fillna(0)
    
#     เรียงลำดับข้อมูลโดยความรุนแรงของการชนกัน
    pg3=pg2.sort_values(by=['SEVERITYCODE_1', 'SEVERITYCODE_2'], ascending=False)    
    pg3 =pg3.head(15)
    dfg1=list(pg3[param])
    ser1=list(pg3['SEVERITYCODE_1'])
    ser2=list(pg3['SEVERITYCODE_2'])
#     print(ser2)
    return dfg1,ser1,ser2
```

```python
def bar_graph(param,wi):
    ser_1=param_graph(param)[1]
    ser_2=param_graph(param)[2]
    label=param_graph(param)[0]
    ind = np.arange(len(label)) 
    width = 0.35  
    fig, ax = plt.subplots(figsize=(wi,6))
    rects1=ax.bar(ind, ser_1, width, label='1')
    rects2=ax.bar(ind + width, ser_2, width,
        label='2')

    ax.set_ylabel('Count')
    ax.set_xlabel(param)
    ax.set_title(f'{param}'+' by Serverity')
    ax.set_xticklabels(label)
    ax.set_xticks(ind + width / 2)
    ax.legend(loc='best')
    def autolabel(rects):
        """Attach a text label above each bar in *rects*, displaying its height."""
        for rect in rects:
            height = rect.get_height()
            plt.annotate('{}'.format(height),
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom')


    autolabel(rects1)
    autolabel(rects2)
    fig.tight_layout()

    plt.show()
```

```python
# กราฟแสดงประเภทสถานที่ชนกัน
bar_graph('ADDRTYPE',8)
```

จากกราฟพบว่าสถานที่ชนกันมากที่สุดคือ ถนน (Block) , สี่แยก(Intersection) และซอย(Alley) ตามลำดับโดยพบว่าจำนวนการชนกันที่มีความรุนแรงระดับที่ 1 มากกว่าความรุนแรงระดับที่ 2 ทั้งสามสถานที่

```python
# กราฟแสดงลักษณะการชนกัน
bar_graph('COLLISIONTYPE',16)
```
